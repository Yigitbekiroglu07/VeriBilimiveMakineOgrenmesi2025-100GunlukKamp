```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_list, "r") # y --> yellow, g --> green
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
# numpy
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
np_age_list
```

```python
np_weight_list
```

```python
plt.plot(np_age_list, np_weight_list, "g")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age vs height")
plt.show()
```

```python
np_arr1 = np.linspace(0,10,20)
```

```python
np_arr1
```

```python
np_arr2 = np_arr1 ** 3
```

```python
np_arr2
```

```python
plt.plot(np_arr1, np_arr2)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(np_arr1,np_arr2,"r*")
plt.subplot(1,2,2)
plt.plot(np_arr2,np_arr1,"b--")
plt.show()
```

```python
my_figure = plt.figure()
my_figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
my_figure_axes.plot(np_arr1, np_arr2, "g")
my_figure_axes.set_xlabel("X Axis")
my_figure_axes.set_ylabel("Y Axis")
my_figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_figure = plt.figure(dpi = 100)

new_axes = new_figure.add_axes([0.1,0.1,0.5,0.5])

new_axes.plot(np_arr1, np_arr1**2, label = "numpy array **2")
new_axes.plot(np_arr1, np_arr1**3, label = "numpy array **3")
new_axes.legend()

plt.show()
```
